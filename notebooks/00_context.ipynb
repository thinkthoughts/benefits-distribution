# 00 — Connected Lanes Specify Benefit Distribution

**Repository:** `benefits-distribution`  
**Engineering statement:** **Benefit Distribution Specifies Sustainable Development**

Notebook 00 defines the language of the repo.

It begins with the mod-30 constraint:

\[
30 = 2\times3\times5
\]

\[
\{1,7,11,13,17,19,23,29\}
\]

Then it extends that connected-lane structure to deployment systems:

\[
\pi_t \rightarrow T_t \rightarrow d_t \rightarrow I_t
\]

where \(\pi_t\) is deployment policy, \(T_t\) is telemetry, \(d_t\) is benefit distribution, and \(I_t\) is sustainable-development indicators.

> Leading constraints specify connected lanes. Connected lanes specify benefit distribution. Benefit distribution specifies sustainable development.

## 1. Setup

Notebook 00 uses no synthetic datasets, no arbitrary telemetry matrices, and no example edge weights.

It defines three figures:

1. **Connected Lanes**
2. **Leading Specification**
3. **Repository Architecture**

The later notebooks add telemetry, estimation, distribution graphs, dynamics, optimization, controllers, and benchmarks.

```python
from pathlib import Path
from zipfile import ZipFile
import numpy as np
import matplotlib.pyplot as plt

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

FIGURES = ROOT / "figures"
RESULTS = ROOT / "results"
DATA = ROOT / "data"

for folder in [FIGURES, RESULTS, DATA]:
    folder.mkdir(parents=True, exist_ok=True)

def savefig(name):
    path = FIGURES / name
    plt.savefig(path, dpi=240, bbox_inches="tight")
    print("saved:", path)
    return path

print("ROOT:", ROOT)
```

## 2. Figure 1 — Connected Lanes

\[
30=2\times3\times5
\]

\[
\{1,7,11,13,17,19,23,29\}
\]

\[
\pi_t \rightarrow T_t \rightarrow d_t \rightarrow I_t
\]

Constraint specifies connected lanes across arithmetic and deployment systems.

```python
residues = [1, 7, 11, 13, 17, 19, 23, 29]

plt.figure(figsize=(13, 6))
ax = plt.gca()
ax.axis("off")

ax.text(0.5, 0.95, "Connected Lanes", ha="center", va="center", fontsize=20, fontweight="bold")

ax.text(0.25, 0.78, r"$30 = 2 \times 3 \times 5$", ha="center", fontsize=22)
ax.text(0.25, 0.65, "constraint", ha="center", fontsize=11)
ax.annotate("", xy=(0.25, 0.51), xytext=(0.25, 0.61), arrowprops=dict(arrowstyle="->", linewidth=2))

xs = np.linspace(0.09, 0.41, len(residues))
for x, r in zip(xs, residues):
    ax.add_patch(plt.Circle((x, 0.43), 0.035, fill=False, linewidth=1.8))
    ax.text(x, 0.43, str(r), ha="center", va="center", fontsize=12)
ax.text(0.25, 0.31, "connected arithmetic lanes", ha="center", fontsize=12, fontweight="bold")

labels = [r"$\pi_t$", r"$T_t$", r"$d_t$", r"$I_t$"]
subs = ["policy", "telemetry", "distribution", "indicators"]
x2 = np.linspace(0.60, 0.90, len(labels))
y2 = 0.60
for i, (x, lab, sub) in enumerate(zip(x2, labels, subs)):
    ax.add_patch(plt.Rectangle((x-0.045, y2-0.05), 0.09, 0.10, fill=False, linewidth=1.8))
    ax.text(x, y2+0.012, lab, ha="center", va="center", fontsize=18)
    ax.text(x, y2-0.09, sub, ha="center", va="center", fontsize=10)
    if i < len(labels)-1:
        ax.annotate("", xy=(x2[i+1]-0.055, y2), xytext=(x+0.055, y2), arrowprops=dict(arrowstyle="->", linewidth=2))

ax.text(0.75, 0.31, "connected deployment lanes", ha="center", fontsize=12, fontweight="bold")
ax.plot([0.5, 0.5], [0.22, 0.84], linewidth=1)
ax.text(0.5, 0.09, "Constraint specifies connected lanes across arithmetic and deployment systems.", ha="center", fontsize=13)

savefig("00_connected_lanes.png")
plt.show()
```

## 3. Figure 2 — Leading Specification

\[
\text{Constraint}
\rightarrow
\text{Connected lanes}
\rightarrow
\text{Deployment}
\rightarrow
\text{Benefit distribution}
\rightarrow
\text{Sustainable development}
\]

Leading constraints specify deployment structure. Sustainable development follows benefit distribution.

```python
plt.figure(figsize=(8, 9))
ax = plt.gca()
ax.axis("off")

steps = ["Constraint", "Connected lanes", "Deployment", "Benefit distribution", "Sustainable development"]
ys = np.linspace(0.82, 0.20, len(steps))
x = 0.46

for i, (step, y) in enumerate(zip(steps, ys)):
    width = 0.44 if len(step) < 18 else 0.58
    ax.add_patch(plt.Rectangle((x-width/2, y-0.045), width, 0.09, fill=False, linewidth=1.8))
    ax.text(x, y, step, ha="center", va="center", fontsize=14, fontweight="bold" if i in [0,1,3] else None)
    if i < len(steps)-1:
        ax.annotate("", xy=(x, ys[i+1]+0.055), xytext=(x, y-0.055), arrowprops=dict(arrowstyle="->", linewidth=2))

side = ["leading", "specifies", "deploys", "distributes", "follows"]
for word, y in zip(side, ys):
    ax.text(0.84, y, word, ha="center", va="center", fontsize=11)

ax.text(0.5, 0.94, "Leading Specification", ha="center", fontsize=20, fontweight="bold")
ax.text(0.5, 0.07, "Leading constraints specify deployment structure. Sustainable development follows benefit distribution.", ha="center", fontsize=11)

savefig("00_leading_specification.png")
plt.show()
```

## 4. Figure 3 — Repository Architecture

Each notebook refines the connected deployment lane from constraint to benchmark.

```python
plt.figure(figsize=(13, 4.8))
ax = plt.gca()
ax.axis("off")

stages = [
    ("00", "Connected\nlanes", "Constraint"),
    ("07", "Telemetry", "Telemetry"),
    ("13", "State\nestimation", "State estimation"),
    ("17", "Distribution", "Distribution"),
    ("23", "Dynamics", "Dynamics"),
    ("29", "Optimization", "Optimization"),
    ("37", "Controllers", "Controllers"),
    ("43", "Benchmarks", "Benchmarks"),
]

xs = np.linspace(0.07, 0.93, len(stages))
y = 0.54

for i, (num, title, top) in enumerate(stages):
    ax.add_patch(plt.Rectangle((xs[i]-0.055, y-0.10), 0.11, 0.20, fill=False, linewidth=1.8))
    ax.text(xs[i], y+0.045, num, ha="center", va="center", fontsize=13, fontweight="bold")
    ax.text(xs[i], y-0.035, title, ha="center", va="center", fontsize=9)
    ax.text(xs[i], 0.82, top, ha="center", va="center", fontsize=10)
    if i < len(stages)-1:
        ax.annotate("", xy=(xs[i+1]-0.065, y), xytext=(xs[i]+0.065, y), arrowprops=dict(arrowstyle="->", linewidth=1.8))

ax.text(0.5, 0.95, "Repository Architecture", ha="center", fontsize=20, fontweight="bold")
ax.text(0.5, 0.18, "Each notebook refines the connected deployment lane from constraint to benchmark.", ha="center", fontsize=12)

savefig("00_repository_architecture.png")
plt.show()
```

## 6. Final download cell

Run the final cell after generating figures. In Google Colab, it downloads the generated outputs zip automatically.

```python
from pathlib import Path
from zipfile import ZipFile

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

ZIP = ROOT / "benefits-distribution-00-context-outputs.zip"

with ZipFile(ZIP, "w") as z:
    for folder_name in ["figures", "results", "data"]:
        folder = ROOT / folder_name
        if folder.exists():
            for file in folder.rglob("*"):
                if file.is_file():
                    z.write(file, file.relative_to(ROOT))

print("Created:", ZIP)

try:
    from google.colab import files
    files.download(str(ZIP))
except Exception:
    try:
        from IPython.display import FileLink, display
        display(FileLink(str(ZIP)))
    except Exception:
        print("Download manually:", ZIP)
```
